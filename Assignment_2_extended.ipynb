# check if a number is even or odd where number is taken as input

```python
numb = input("Enter integer number: ")
if numb.isdigit():
    numb = int(numb)
    if numb%2 == 0:
        print(f"{numb} is Even")
    else:
        print(f"{numb} is Odd")
else:
    print("Invalid input!")
```

# User se marks lo aur grade do:

# 90+ → A
# 80+ → B
# 70+ → C
# 60+ → D
# warna → Fail

```python
sub1 = float(input("Enter subject one marks: "))
sub2 = float(input("Enter subject two marks: "))
sub3 = float(input("Enter subject three marks: "))
sub4 = float(input("Enter subject four marks: "))
sub5 = float(input("Enter subject five marks: "))
obtain_marks = sub1 + sub2 + sub3 + sub4 + sub5
percent = (obtain_marks/500)*100
grade = ""
if percent >= 90:
    grade = "A" 
elif percent >= 80:
    grade = "B" 
elif percent >= 70:
    grade = "C" 
elif percent >= 60:
    grade = "D" 
else:
    grade = "F"

print(f'''
--------------- Result ---------------
    Obtain Marks  : {obtain_marks}
    Total Marks   : 500
    Percentage    : {percent}
    Grade         : {grade}
    Status        : {"You are failed" if grade == "F" else "You are passed"}
--------------------------------------
''')
```

# User se day number (1–7) lo:

# 1 → Monday
# 2 → Tuesday
## ...
# 7 → Sunday

```python
day = int(float(input("Enter day number")))
if day == 1:
    print("Monday")
elif day == 2:
    print("Tuesday")
elif day == 3:
    print("Wednesday")
elif day == 4:
    print("Thursday")
elif day == 5:
    print("Friday")
elif day == 6:
    print("Saturday")
elif day == 7:
    print("Sunday")
else:
    print("Invalid day number!")
```

# User se month number lo:

# 12,1,2 → Winter
# 3,4,5 → Spring
# 6,7,8 → Summer
# warna → Autumn

```python
month = int(float(input("Enter month number")))
if month == 1 or month == 2 or month == 12:
    print("Winter")
elif month == 3 or month == 4 or month == 5:
    print("Spring")
elif month == 6 or month == 7 or month == 8:
    print("Summer")
else:
    print("Autumn")
```

# Simple calculator:
# User se operator lo (+, -, *, /):

# → addition
# → subtraction
# → multiplication
# / → division
# warna → "Invalid operator"

```python
op = input("Enter operator number")
if op == '+':
    print("Addition")
elif op == '-':
    print("Subtraction")
elif op == '*':
    print("Multiplication")
elif op == '/':
    print("Division")
else:
    print("Invalid operator")
```
